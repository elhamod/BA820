##EDA and Preprocessing

```python
# drop adult movies
df_movies = df_movies[df_movies['adult'] == 'False']
# drop movies not released
df_movies = df_movies[df_movies['status'] == 'Released']
# keep only original langugae is English
df_movies = df_movies[df_movies['original_language'] == 'en']
# convert id and budget to int
df_movies['id'] = df_movies['id'].astype(int)
df_movies['budget'] = df_movies['budget'].astype(int)
# drop movies with low or unknown budget
df_movies = df_movies[df_movies['budget'] > 100000]
# drop movies where revenue or runtime is exactly zero (probably an error)
df_movies = df_movies[df_movies['revenue'] != 0.0]
df_movies = df_movies[df_movies['runtime'] != 0.0]
# drop nans
df_movies = df_movies.dropna(subset=["budget", "overview" , "popularity", "release_date", "title"])
# drop some columns
df_movies = df_movies.drop(columns=['adult', 'homepage', 'imdb_id', 'tagline', 'status', 'original_language' ,'spoken_languages', 'video', 'original_title', 'poster_path'])
#  add a column for whether a movie belongs to a collection
df_movies['belongs_to_collection'] = df_movies['belongs_to_collection'].notna()
# drop duplicates
df_movies = df_movies.drop_duplicates(subset=['id'])
# treat id as dataframe id
df_movies = df_movies.set_index('id')

df_movies
```

```python
# remove from credits all movies not in df_movies
df_credits = df_credits[df_credits['id'].isin(df_movies.index)]
# drop nans
df_credits = df_credits.dropna()
# drop duplicates
df_credits = df_credits.drop_duplicates(subset=['id'])
# treat id as dataframe id
df_credits = df_credits.set_index('id')

df_credits
```

```python
# remove from keywords and ratinds all movies not in df_movies
df_keywords = df_keywords[df_keywords['id'].isin(df_movies.index)]
df_ratings = df_ratings[df_ratings['movieId'].isin(df_movies.index)]
# drop nans
df_keywords = df_keywords.dropna()
df_ratings = df_ratings.dropna()
# drop duplicates
df_keywords = df_keywords.drop_duplicates(subset=['id'])
df_ratings = df_ratings.drop_duplicates()
# treat id as dataframe id
df_keywords = df_keywords.set_index('id')

display(df_keywords)
display(df_ratings)
```
