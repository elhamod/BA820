```python
from datasets import load_dataset

# ds = load_dataset("textminr/cmu-book-summaries")
ds = load_dataset("Iliasselyaa/CMUBookSummaryDataset")

#get dataframe
df = ds["train"].to_pandas()

# drop all ids
df = df.drop(columns=["id", "wikipedia_id", "freebase_id"])



df
```

##EDA and Preprocessing

```python
#remove Nans
df = df.dropna().reset_index(drop=True)

# extract year from publication_date as integer
df["publication_date"] = df["publication_date"].str[:4].astype(int)

# make genres as a list
df["genres"] = df["genres"].str.split(",")

display(df)

# get all unique genres with their counts
genres = df["genres"].explode().value_counts()

genres
```
